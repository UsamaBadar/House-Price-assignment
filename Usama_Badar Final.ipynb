```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Read the CSV file inside the extracted folder into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/code/houseprices.csv')
# Print the first 5 rows of the DataFrame
df.head()
```

```python
def Bar_Plot(column_name):


  """Setting the size of figure and plotting the Bar plot .
  This function will receive the name of column of which we want to plot the barplot
  Name of categories on x-axis and the value of frequency of category will be on y-axis"""

  plt.figure(figsize = (12, 5))
  # Coverting into dictionary after taking the value count with respect to category
  dict_col = dict(df[column_name].value_counts())
  # Separating the categories and its values for plotting bar plot
  categories = dict_col.keys()
  values = dict_col.values()

  # Create a bar plot
  plt.bar(categories, values)
  # Adding x and y label across their axis
  plt.xlabel(f'{column_name}')
  plt.ylabel('Freqency')
  

  # Add title
  plt.title(column_name)
  # Rotate the x-axis labels by 90 degrees
  plt.xticks(rotation = 90)
  # Show the plot
  plt.show()
 
```

```python
def Pie_Plot(column_name):


  """Pie chart represents the percentage of each category in slices.
  This function will receive the name of column of which we want to plot the pie chart
  """
  
  # Setting the size of figure plotting the pie chart 
  plt.figure(figsize = (12, 5))
  
  # Coverting into dictionary after taking the value count with respect to category
  dict_col = dict(df[column_name].value_counts())

  # Separating the categories and its values for plotting bar plot
  categories = dict_col.keys()
  values = dict_col.values()
  
  plt.pie(values, labels = categories, autopct = '%1.1f%%')
  legends = ['%s' % (l) for l, s in zip(categories, values)]
  
  # Adding legend with percentage and category labels
  plt.legend(labels = legends, bbox_to_anchor = (0.8, 1))

  # Add title
  plt.title(column_name)

  # Rotate the x-axis labels by 90 degrees
  plt.xticks(rotation = 90)
  
  # Show the plot
  plt.show()

```

```python
def LinePlot():


  """Plotting the line plot of every category of GeoName and showing the Average price of every category with respect to Month n Year"""
 
  # Setting the size of figure plotting the pie chart 
  plt.figure(figsize = (16, 9))

  grouped = df.groupby('GeoName')

  # ploting the line plot of each location

  for name, group in grouped:
      plt.plot(group['Month'], group['AveragePrice'], label = name)

  # Add axis labels and title
  plt.xlabel('Month n Year')
  plt.ylabel('Average Price')
  plt.title('Average Price by Type')

  # Add a legend
  plt.legend()

  # Show the plot
  plt.show() 

```

```python
def main(): 


  """this is the main function where we are calling the other 3 fundciotns which we have defined"""
  Bar_Plot("GeoName")
  Pie_Plot("GeoEntityName")
  LinePlot()
        


if __name__=="__main__":


  """Python Interpreter will directly come here and will run these lines"""
  main()

  
```
